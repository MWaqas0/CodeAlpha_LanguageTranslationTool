```python
from deep_translator import GoogleTranslator
import gradio as gr

def translate_text(text, source_lang, target_lang):
    translated = GoogleTranslator(
        source=source_lang,
        target=target_lang
    ).translate(text)
    return translated

iface = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(label="Enter Text"),
        gr.Dropdown(["auto","en","ur","hi","fr","de"], label="Source Language"),
        gr.Dropdown(["en","ur","hi","fr","de"], label="Target Language")
    ],
    outputs="text",
    title="Language Translation Tool"
)

iface.launch()
```
